###Desafio 01: Investique porque a classe tratamento tratamento é tão desbalanceada.

```python
#Primeiramente entender o que é o "tratamento".
dados['tratamento']
```

```python
#Então compreender quais os possíveis valores em tratamento
dados['tratamento'].unique()
```

```python
#Dando mais um pequeno passo na direção desejada, entender melhor essa distribuição
dados['tratamento'].value_counts()
```

```python
#Entender a distribuição dos tratamentos com droga e com controle de forma normalizada
dados['tratamento'].value_counts(normalize = True)
```

```python
#Afim de uma visualização mais amigável, um gráfico de pizza é interessante para entender a discrepância entre os tratamentos.
dados['tratamento'].value_counts().plot.pie()
```

```python
#Partindo disso, visualizando as drogas analisadas no estudo.
dados['droga'].value_counts()
```

```python
#Visualizando os dados que estão com controle afim de entender melhor os dados.
dados[dados['tratamento']=='com_controle']
```

```python
#Aparentemente há uma relação entre a droga 'cacb2b860' e o tratamento 'com controle'. Então fazer se há uma igualdade entre a quantidade de ocorrências.
dados['droga'].value_counts()['cacb2b860'] == dados['tratamento'].value_counts()['com_controle']
```

```python
#Afim de confirmar, mostrar que há igualdade entre os dados de tratamento com controle e da droga 'cacb2b860'.
dados[dados['tratamento']=='com_controle'] == dados[dados['droga']=='cacb2b860']
```

Assim sendo, a classe tratamento é desbalanceada porque o objetivo é fazer a comparação com os diferentes tipos de droga. Ademais, o tratamento com a droga `'cacb2b860'` é o controle. Talvez esse seja o farmaco padrão já conhecido.

###Desafio 03: Proporção das classes tratamento.



```python
#Conforme analisado no Desafio 01, temos o gráfico de pizza para melhor compreensão.
dados['tratamento'].value_counts().plot.pie()
```

```python
#Agora deixando o gráfico de forma mais amigável, obtemos:
import matplotlib.pyplot as plt
dados['tratamento'].value_counts(normalize=True)
labels = ['Com Droga', 'Com Controle']
titulos = dados['tratamento'].value_counts()
explode = (0,0) #Sem explosão alguma
cores = ['yellow', 'blue']

plt.pie(titulos, explode=explode, labels=labels, colors=cores, autopct= '%1.2f%%', shadow=True, startangle=1)
plt.title ("Comparação entre Tratamentos", fontsize = '20')
plt.axis('equal')
plt.legend()
plt.show()
```

###Desafio 07: Deixar os gráficos mais elegantes (Matplotlib.pyplot).

O Desafio 03 já fez uso desses gráficos de forma mais elegante. Todavia, afim de exemplificar melhor, irei fazer o gráfico de pizza anteriormente descrito com explosão e gráfico de barras para o tempo.

```python
#Gráfico de Pizza atualizado
dados['tratamento'].value_counts(normalize=True)
labels = ['Com Droga', 'Com Controle']
titulos = dados['tratamento'].value_counts()
explode = (0,0.2) #Sem explosão alguma
cores = ['yellow', 'blue']

plt.pie(titulos, explode=explode, labels=labels, colors=cores, autopct= '%1.2f%%', shadow=True, startangle=1)
plt.title ("Comparação entre Tratamentos", fontsize = '20')
plt.axis('equal')
plt.legend()
plt.show()
```

```python
#Agora gráficos de barras para o tempo de forma mais imediata.
dados['tempo'].value_counts().plot.bar()
plt.title('Tempo em horas', size = 20)
plt.xlabel('Tempo')
plt.ylabel('Quantidade de experimentos')
plt.xticks(rotation = 0)
plt.show()
```
